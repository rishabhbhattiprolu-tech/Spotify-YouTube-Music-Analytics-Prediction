```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = pd.read_csv("Spotify_Youtube.csv")
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.drop(["Url_spotify", "Track", "Url_youtube", "Description", "Unnamed: 0", "Title", "Artist", "Album", "Channel", "Uri"], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Album_type"].value_counts()
```

```python
df["Album_type"].replace({"album": 0, "single": 1, "compilation": 2}, inplace=True)
df["Licensed"].replace({False: 0, True: 1}, inplace=True)
df["official_video"].replace({False: 0, True: 1}, inplace=True)
```

```python
df.head()
```

```python
X = df.drop("Likes", axis=1)
y = df["Likes"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
y_pred = lm.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
print(f"MAE: {mae}, MSE: {mse}, RMSE: {rmse}, R2: {r2}")
```
